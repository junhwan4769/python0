```python
def add_cm(height):
    if isinstance(height, int):
        return f'{height}cm'
    else: 
        return int(height.replace('cm', ''))
```

```python
print(add_cm(165))
```

```python
print(add_cm('165cm'))
```

```python
df7 = pd.read_csv('data/score.csv')
df7.set_index('지원번호', inplace = True)
df7
```

```python
df7_score = df7.loc[:, '국어':'사회']
df7_score
```

```python
import numpy as np
```

```python
df7_score['총점'] = df7_score.apply(lambda row:sum(row['국어':'사회']), axis=1)
df7_score['평균'] = df7_score.apply(lambda row:row['총점']/5, axis=1)
df7_score
```

```python
df7_score.loc['평균'] = df7_score.apply(np.mean, axis=0)
df7_score
```
