```python
import pandas as pd
```

```python
cou = pd.read_csv('강좌.csv')
cou.tail()
```

```python
enroll = pd.read_csv('수강.csv')
enroll.head()
```

```python
group = enroll.groupby('강좌번호').size().to_frame('수강인원')
group
```

```python
group.reset_index(inplace=True)
group
```

```python
merge = pd.merge(cou, group, how='left', on='강좌번호')
merge
```

```python
merge.info()
```

```python
merge.fillna({'수강인원':0}, inplace=True)
```

```python
merge['수강인원'] = merge['수강인원'].astype(int)
merge
```

```python
def stu_list():
    enroll = pd.read_csv('수강.csv')
    group = enroll.groupby('학생번호').size().to_frame('강좌수')
    stu = pd.read_csv('학생.csv')
    merge = stu.merge(group, how='left', on='학생번호')
    merge.fillna({'강좌수':0}, inplace = True)
    merge.set_index('학생번호', inplace = True) 
    merge['강좌수']=merge['강좌수'].astype(int)
    return merge
```

```python
stu_list()
```
