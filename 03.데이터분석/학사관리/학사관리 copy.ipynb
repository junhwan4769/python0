```python
import pandas as pd
```

```python
prof = pd.read_csv('교수.csv', index_col='교수번호')
prof
```

```python
stu = pd.read_csv('학생.csv', index_col='학생번호')
stu
```

```python
filt = stu['지도교수'] == 221
stu = stu[filt]
```

```python
prof2 = prof.copy()
prof2.reset_index(inplace=True)
prof2 = prof2[['교수번호', '교수이름', '교수학과']]
prof2
```

```python
prof_stu = stu.merge(prof2, how='inner', left_on= '지도교수', right_on='교수번호')
prof_stu
```

```python
prof_stu.drop(columns='교수번호', inplace=True)
prof_stu
```

```python
prof_stu.sort_values('지도교수')
prof_stu
```

```python
prof_stu.groupby('지도교수').size()
```

```python
for code in prof.index:
    df = prof_stu.groupby('지도교수').get_group(code)
    print(df)
```

```python
for idx in prof.index:
    row = prof.loc[idx]
    print(f'교수번호:{idx}, 교수이름:{row["교수이름"]} 교수학과:{row["교수학과"]}')
    print('지도학생목록-----------------------------')
    filt = stu['지도교수'] == idx
    df = stu[filt]
    for idx in df.index:
        row = df.loc[idx]
        print(f'학생번호:{idx}, 학생이름:{row["학생이름"]} 학생학과:{row["학생학과"]}')
```
