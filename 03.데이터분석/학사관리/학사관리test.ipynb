```python
import pandas as pd
```

```python
cou = pd.read_csv('강좌.csv')
cou = cou[['강좌번호', '강좌이름', '강의시수', '담당교수']]
cou.head()
```

```python
prof = pd.read_csv('교수.csv')
prof = prof[['교수번호', '교수이름']]
prof.head()
```

```python
cou_prof = pd.merge(cou, prof, left_on='담당교수', right_on='교수번호')
cou_prof.head()
```

```python
cou_prof.drop(columns='교수번호')
cou_prof
```

```python
enroll = pd.read_csv('수강.csv')
group = enroll.groupby('강좌번호')['점수'].mean().to_frame()
group.reset_index(inplace=True)
group.head()
```

```python
group_cou_prof = group.merge(cou_prof, on='강좌번호')
group_cou_prof
```

```python
group_cou_prof.drop(columns='교수번호', inplace=True)
group_cou_prof
```

```python
result = group_cou_prof.copy()
result = result.iloc[:, [0,2,5,3,1]]
result.set_index('강좌번호', inplace=True)
result
```
