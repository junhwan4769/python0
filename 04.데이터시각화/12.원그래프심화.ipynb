```python
import pandas as pd
enroll = pd.read_csv('data/수강.csv')
enroll
```

```python
group = enroll.groupby('강좌번호').size().to_frame('인원수')
group
```

```python
cou = pd.read_csv('data/강좌.csv')
cou
```

```python
merge = cou.merge(group, on= '강좌번호')
merge
```

```python
labels = merge['강좌이름']
values = merge['인원수']
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, wedgeprops={'width':0.8,'edgecolor':'black', 'linewidth':2}, colors=colors,
        autopct='%.1f%%', explode=[0.05]*len(values))
plt.legend(loc=(1.3,0.5))
plt.show()
```

```python
stu = pd.read_csv('data/학생.csv')
stu
```

```python
group = stu.groupby('학생학과').size()
group
```

```python
labels = group.index
values = group.values
plt.pie(values, labels=labels, pctdistance= 0.7, wedgeprops={'width':0.5,'edgecolor':'white', 'linewidth':2},
        autopct='%.1f%%')
plt.legend(loc=(1.3,0.5))
plt.show()
```

```python
def grade(score):
    if score >=90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
```

```python
enroll = pd.read_csv('data/수강.csv')
enroll
```

```python
enroll['학점'] = enroll['점수'].apply(grade)
enroll
```

```python
group = enroll.groupby('학점').size()
group
```

```python
plt.pie(group.values, labels=group.index + '학점')
plt.show()
```
