```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)

import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
def grade(score):
    if score >=90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
```

```python
enroll = pd.read_csv('data/수강.csv')
enroll
```

```python
enroll['학점'] = enroll['점수'].apply(grade)
enroll
```

```python
group = enroll.groupby('학점').size()
group
```

```python
plt.pie(group.values, labels=group.index + '학점',wedgeprops={'width':0.7,'edgecolor':'white', 'linewidth':2},
        autopct='%.1f%%', textprops={'size':15})
plt.show()
```
