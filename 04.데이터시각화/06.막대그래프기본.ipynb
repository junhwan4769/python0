```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)
```

```python
name = df['이름']
height = df['키']
```

```python
colors = ['gray', 'navy', 'black', 'green', 'red', 'yellow', 'blue', 'orange']
plt.bar(name, height, color=colors, alpha=0.2)
plt.show()
```

```python
plt.bar(name, height, color= 'black', width=0.5)
plt.ylim(160,220)
plt.yticks([y for y in range(160, 221, 5)])
plt.xticks(name, rotation = 90)
plt.show()
```
