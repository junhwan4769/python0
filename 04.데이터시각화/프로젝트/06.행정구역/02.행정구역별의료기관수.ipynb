```python
filt = local['시도'] == '인천광역시'
local1 = local[filt]
local1.head()
```

```python
def custom_autopct(pct):
    if pct >=3:
        return f'{pct:.0f}%'
    else:
        ''
    #return f''{pct:.0f}%' if pct >=3 else ''
```

```python
values = local1['count']
labels = local1['군구']
plt.pie(values, labels=labels, autopct=custom_autopct, rotatelabels=False,
        explode=[0.05]*len(local1), wedgeprops={'width':0.9, 'alpha':0.5, 'edgecolor':'black'},
        textprops={'fontsize':8})
plt.show()
```
