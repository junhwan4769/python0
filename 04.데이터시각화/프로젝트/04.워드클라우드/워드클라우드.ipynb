```python
with open('영어가사.txt', 'r') as file:
    text = file.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud().generate(text)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('영어가사.png')
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
heart_mask = np.array(Image.open('heart_mask.png'))
heart_mask
```

```python
wc = WordCloud(
    width=800, 
    height=800, 
    background_color='white',
    min_font_size=20,
    mask=heart_mask).generate(text)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('영어가사하트.png')
plt.show()
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()
```

```python
with open('애국가.txt', 'r', encoding='utf-8') as file:
    text = file.read()
print(text)
```

```python
nouns = okt.nouns(text)
nouns
```

```python
nouns = [noun for noun in nouns if len(noun) > 1]
nouns
```

```python
counter = Counter(nouns)
counter
```

```python
wc = WordCloud(width=800, 
               height=800, 
               font_path='MALGUN.TTF').generate_from_frequencies(counter)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(
    width=800, 
    height=800, 
    background_color='gray',
    min_font_size=20,
    mask=heart_mask, font_path='MALGUN.TTF').generate_from_frequencies(counter)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('애국가하트.png')
plt.show()
```
