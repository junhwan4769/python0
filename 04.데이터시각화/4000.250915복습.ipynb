```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
#학생별(이름) 국어,영어,수학 점수를 꺾은선 그래프
plt.figure(figsize=(10,5))
plt.title('학생별 국어 영어 수학 점수', size = 10)
plt.plot(df['이름'], df['국어'], marker='o', label='국어', markersize=10, 
         markeredgecolor='black', markeredgewidth=3, markerfacecolor='gray')
plt.plot(df['이름'], df['영어'], marker='o', label='영어', markersize=10, 
         linewidth=3, linestyle='--', color= 'green')
plt.plot(df['이름'], df['수학'], marker='o', label='수학', markersize=10, 
         linewidth=2, color= 'navy')
plt.legend(loc='lower right')
plt.ylim(0,110)
yticks = [y for y in range(0, 101, 50)]
plt.yticks(yticks, [f'{y}점' for y in yticks])
plt.xticks(df['이름'], [x+'님' for x in df['이름']])
plt.xlabel('이름', size=15, loc='left', color='navy')
plt.ylabel('점수', size=15, loc='top', color='pink')
plt.show()
```

```python
#학생별 (키) 막대그래프
plt.figure(figsize=(10,3))
plt.bar(df['이름'], df['키'], color='none', hatch='/', edgecolor='black')
plt.ylim(160,210)
yticks = [y for y in range(160, 211, 10)]
plt.yticks(yticks, [f'{y}cm' for y in yticks])
xticks = [f'{df.loc[x,"이름"]}\n{df.loc[x,"지원번호"]}' for x in range(len(df))]
plt.xticks(df['이름'], xticks)
plt.show()
```

```python
#학생별 평균점수 막대그래프
df['평균'] = df.apply(lambda row:row['국어':'사회'].mean(), axis=1)
plt.bar(df['이름'], df['평균'], color='black')
for idx, avg in enumerate(df['평균']):
    plt.text(idx, avg+1, f'{avg:.2f}', ha='center')
plt.show()
```
