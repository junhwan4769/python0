```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)
```

```python
kor = list(df['국어'])
eng = list(df['영어'])
mat = list(df['수학'])
soe = list(df['사회'])
sci = list(df['과학'])
avg = list((df['국어'] + df['영어'] + df['수학'] + df['사회'] + df['과학']) / 5)
name = df['이름']
```

```python
plt.figure(figsize = (10, 5))
plt.ylim(0,100)
plt.ylabel('평균점수')
plt.xlabel('학생이름')
plt.yticks([y for y in range(0, 101, 5)])
plt.title('학생별 평균\n', size = 20)
plt.bar(name, avg, color = 'none', edgecolor = 'black')
for idx, avg in enumerate(avg):
    plt.text(idx, avg+1, avg, ha='center')
plt.show()
```
