```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)

import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
group = df.groupby('SW특기').size()
group
```

```python
labels = group.index
values = group.values
```

```python
values
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=True)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC'] 
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=180, counterclock=True)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=[0.05] * len(values), 
        autopct='%.1f%%', startangle=180, counterclock=True)
plt.legend(loc=(1.3,0.5))
plt.show()
```

```python
def custom_autopct(pct):
    if pct <=2:
        return ''
    else:
        return f'{pct:.0f}%'
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, explode=[0.05] * len(values), colors=colors,
        wedgeprops={'width':0.8, 'edgecolor':'black', 'linewidth':2},
        autopct='%.1f%%', startangle=180, counterclock=True)
plt.legend(loc=(1.3,0.5))
plt.show()
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, explode=[0.05] * len(values), colors=colors,
        wedgeprops={'width':0.8, 'edgecolor':'black', 'linewidth':2},
        autopct=custom_autopct, startangle=180, counterclock=True)
plt.legend(loc=(1.3,0.5))
plt.show()
```
