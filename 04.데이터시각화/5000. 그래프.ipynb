```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)

import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
plt.figure(figsize=(10,5))
plt.bar(df['이름'],df['국어'], color='black', label='국어')
plt.bar(df['이름'],df['영어'], bottom=df['국어'], color='gray', label='영어')
plt.bar(df['이름'],df['수학'], bottom=df['국어'] + df['영어'], color='navy', label='수학')
plt.ylim(0,300)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)

import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
x = df.index * 4 
plt.figure(figsize=(10,5))
plt.bar(x, df['국어'], width=1, label='국어', color='black')
plt.bar(x+1, df['영어'], width=1, label='영어', color='gray')
plt.bar(x+2, df['수학'], width=1, label='수학', color='navy')
plt.legend()
xticks = [xtick+1 for xtick in x]
plt.xticks(xticks, df['이름'])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)

import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
plt.scatter(df['국어'], df['영어'], sizes=df['학년']*500,
            c=df['학년'], alpha=0.3)
plt.colorbar(ticks=[1,2,3], orientation='horizontal', label='학년', shrink=0.5)
for idx in range(len(df)):
    x = df.loc[idx, '국어']
    y = df.loc[idx, '영어']
    name = df.loc[idx, '이름']
    plt.text(x, y, name, ha='center', size=6)
plt.ylim(0,120)
plt.show()
```
