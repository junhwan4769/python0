```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)
```

```python
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
name = list(df['이름'])
height = list(df['키'])
name, height
```

```python
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름')
plt.ylabel('키(cm)')
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, linewidth = '3', marker = 's')
plt.show()
```

```python
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름')
plt.ylabel('키(cm)')
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, linewidth = '3', marker = 'o', markersize = 10, markerfacecolor='gray', 
         markeredgecolor = 'black', markeredgewidth = 3)
plt.show()
```

```python
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름')
plt.ylabel('키(cm)')
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, linewidth = '3', linestyle = ':', color = 'green',
         marker = 'o', markersize = 10, markerfacecolor='gray', 
         markeredgecolor = 'black', markeredgewidth = 3)
plt.show()
```

```python
plt.title('학생별 신장\n', size = 20)
plt.xlabel('이름')
plt.ylabel('키(cm)')
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height,
         lw = 3, color='black', ls='--', marker= 'o', ms = 10,
         mfc = 'gray', mec = 'red', mew = 3)
plt.show()
```

```python
plt.figure(figsize=(10 ,3))
plt.ylim(160, 220)
plt.plot(name, height, marker = 'o')
for idx, txt in enumerate(height):
    plt.text(name[idx], height[idx]+3, f'{txt}cm', size = 10, ha= 'center')
plt.show()
```
