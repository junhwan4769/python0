```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('data/LinearRegressionData.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values #독립변수
y = dataset.iloc[:, -1].values #종속변수
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
plt.scatter(X,y, label='실제점수', color='black')
plt.plot(X, reg.predict(X), label='예측점수', color='green')
xticks = [int(x) for x in range(len(X))]
plt.xticks =(xticks)
plt.grid(True, ls='--', lw=0.5)
plt.legend()
plt.xlabel('공부시간')
plt.ylabel('시험점수')
plt.show()
```
